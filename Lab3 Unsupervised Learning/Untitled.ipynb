```python
def missing_value_count(train_set):
    rows,N = train_set.shape
    missing_value = train_set.isnull().sum()
    for i in range(N):
        if missing_value[i] > 0:
            miss_persent = missing_value[i]/rows*100
            print(missing_value[i],f":\t %.2f" % miss_persent,"% \t",train_set.columns[i])
    
    print(train_set.shape)
```

```python
def fill_missing_values(train_set,test_set):
    for i in train_set.columns:
        if train_set[i].isnull().sum() > 0:    
            temp_fill_value = train_set[i].value_counts().index[0]
            train_set[i].fillna(value=temp_fill_value, inplace=True)
    for i in test_set.columns:
        if test_set[i].isnull().sum() > 0:    
            temp_fill_value = train_set[i].value_counts().index[0]
            test_set[i].fillna(value=temp_fill_value, inplace=True)
```

```python
#Створюємо категоріальні критерії
categorical_features = {'Survived','Pclass','Sex','Embarked'}
categorical_features_test = {'Pclass','Sex','Embarked'}
numerical_features = {'Age','SibSp','Parch','Fare'}
for feature in categorical_features:    
    train_set[feature] = train_set[feature].astype("category")
    
for feature in categorical_features_test:
    test_set[feature] = test_set[feature].astype("category")
```

```python
def lable_one_hot_encoding(train_set,test_set):
    le = preprocessing.LabelEncoder()
    train_set = pd.DataFrame(train_set)
    for i in train_set:
        if train_set[i].dtype != ('int64' or 'float64'):
#             print('I am in category',len(list(train_set[i].unique())), end="" )
            if len(list(train_set[i].unique())) == 2:
                train_set[i] = le.fit_transform(train_set[i])
                test_set[i] = le.transform(test_set[i])
#                 print("\t is trasformd")
                print(i)

    train_set = pd.get_dummies(train_set)
    test_set = pd.get_dummies(test_set)
    return train_set , test_set
```

```python
#застосовуємо lable та one_hot_encoding. Перетворюємо категоріалні колонки в чмсловий формат
my_Survived_train = train_set['Survived'] 
train_set.drop(['Survived'], axis='columns', inplace=True)
train_set, test_set = lable_one_hot_encoding(train_set,test_set)
train_set['Survived'] = my_Survived_train
```

```python
def build_correlation_matrix(train_set):    
    corrMatt = train_set.corr()
    mask = np.array(corrMatt)
    mask[np.tril_indices_from(mask)] = False
    fig,ax= plt.subplots()
    fig.set_size_inches(10,10)
    sns.heatmap(corrMatt, cmap="Greens", mask=mask,vmax=.8, square=True,annot=True)
    return corrMatt
```

```python
##Створюємо матрицб кореляції для numerical_train_set
corr_matrix_num = build_correlation_matrix(train_set)
```

```python
#Застосовуємо метод KMeans
from sklearn.cluster import KMeans
plt.figure(figsize=(5, 5))
scaler = StandardScaler()
train_scaled = scaler.fit_transform(train_set)
test_scaled = scaler.fit_transform(test_set)

km = KMeans(n_clusters=2).fit_transform(train_scaled)
tsne = TSNE()

plt.scatter(km[:, 0], km[:, 1])
```

```python
#Застосовуємо метод DBSCAN
scaler = StandardScaler()
train_scaled = scaler.fit_transform(train_set)
test_scaled = scaler.fit_transform(test_set)

dbscan = DBSCAN(eps = 1, min_samples = 2)
tsne = TSNE(perplexity=300.0,)

train_plot = dbscan.fit(train_scaled)

train_result = dbscan.labels_

train_plot_tsne = tsne.fit_transform(train_scaled)
test_plot_tsne = tsne.fit_transform(test_scaled)

plt.scatter(train_plot_tsne[:, 0], train_plot_tsne[:, 1], c=train_result, cmap="jet")
```

```python
#Застосовуємо метод DBSCAN для eps = 3
train_scaled = scaler.fit_transform(train_set)
test_scaled = scaler.fit_transform(test_set)

#change eps
dbscan = DBSCAN(eps = 3, min_samples = 2)
tsne = TSNE(perplexity=300.0,)

train_plot = dbscan.fit(train_scaled)

train_result = dbscan.labels_

train_plot_tsne = tsne.fit_transform(train_scaled)
test_plot_tsne = tsne.fit_transform(test_scaled)

plt.scatter(train_plot_tsne[:, 0], train_plot_tsne[:, 1], c=train_result, cmap="jet")
```

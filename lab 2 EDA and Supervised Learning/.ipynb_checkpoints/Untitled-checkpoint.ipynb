```python
import numpy as np
import math
import csv
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LogisticRegression
import matplotlib
import random
import pandas as pd
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
def missing_value_count(train_set):
    rows,N = train_set.shape
    missing_value = train_set.isnull().sum()
    for i in range(N):
        miss_persent = missing_value[i]/rows*100
        print(i,missing_value[i],f":\t %.2f" % miss_persent,"% \t",train_set.columns[i])
    
    print(train_set.shape)
```

```python
def fill_missing_values(train_set,test_set):
    for i in train_set.columns:
        if train_set[i].isnull().sum() > 0:    
            temp_fill_value = train_set[i].value_counts().index[0]
            train_set[i].fillna(value=temp_fill_value, inplace=True)
    for i in test_set.columns:
        if test_set[i].isnull().sum() > 0:    
            temp_fill_value = train_set[i].value_counts().index[0]
            test_set[i].fillna(value=temp_fill_value, inplace=True)
```

```python
#Deleting columns with missing pencent >= 40%
def delete_columns_miss_over(train_set,test_set,persent):
    missing_value = train_set.isnull().sum()
    N = len(train_set)
    j=0
    for i,elem in enumerate(missing_value):
        temp_miss_persent = elem/N*100
        if temp_miss_persent >= persent:
            test_set.drop(train_set.columns[j], axis='columns', inplace=True)
            train_set.drop(train_set.columns[j], axis='columns', inplace=True)
            j -=1
        j +=1
    print("Now train_set shape is",train_set.shape)
    print("Now test_set shape is",test_set.shape)
```

```python
def create_categorical_features(train_set,test_set):
    for i,column_type in enumerate(train_set.dtypes):
        if ((column_type == 'object') or (('FLAG' or 'LIVE_'  or 'REG_' ) in train_set.columns[i])):
            train_set[train_set.columns[i]] = train_set[train_set.columns[i]].astype('category')            
            test_set[test_set.columns[i]] = test_set[test_set.columns[i]].astype('category')
            print(i,column_type,train_set.columns[i])
```

```python
def finding_col_corr_over(train_set1, corr_matrix, corr):
    train_set = pd.DataFrame(train_set1)
    high_correlation_set = set()
    counter = 0
    df_columns = list(train_set)
    print("Deleting second columns")
    for i in range(len(corr_matrix)):
        for j in range(i + 1, len(corr_matrix)):
            if corr_matrix[df_columns[i]][df_columns[j]] > corr:
#                 print(df_train_numerical_columns[i], ' ', df_train_numerical_columns[j], ' correlation ',  corrMatt[df_train_numerical_columns[i]][df_train_numerical_columns[j]])
                print(df_columns[i], '\t ', df_columns[j], '\t correlation ',  corr_matrix[df_columns[i]][df_columns[j]])

                high_correlation_set.add(df_columns[j])
                counter = counter + 1

    return high_correlation_set
```

```python
def deleting_col_by_corr(train_set,high_correlation_set):
    for feature in high_correlation_set:
        train_set.drop(feature, axis='columns', inplace=True)
    print("Now train_set shape is",train_set.shape)
    return train_set
```

```python
def print_data_feateres(train_set):
    for i, elem in enumerate(train_set.columns):
        print(i, elem)
```

```python
def build_correlation_matrix(train_set):    
    corrMatt = train_set.corr()
    mask = np.array(corrMatt)
    mask[np.tril_indices_from(mask)] = False
    fig,ax= plt.subplots()
    fig.set_size_inches(90,90)
    sns.heatmap(corrMatt, cmap="Greens", mask=mask,vmax=.8, square=True,annot=True)
    return corrMatt
```

```python
def lable_one_hot_encoding(train_set,test_set):
    le = preprocessing.LabelEncoder()
    train_set = pd.DataFrame(train_set)
    for i in train_set:
        if train_set[i].dtype != ('int64' or 'float64'):
#             print('I am in category',len(list(train_set[i].unique())), end="" )
            if len(list(train_set[i].unique())) == 2:
                train_set[i] = le.fit_transform(train_set[i])
                test_set[i] = le.transform(test_set[i])
#                 print("\t is trasformd")
                print(i)

    train_set = pd.get_dummies(train_set)
    test_set = pd.get_dummies(test_set)
    return train_set , test_set
```

```python
updated_train_set['TARGET'] = my_TARGET_train
train_x,test_x,train_y,test_y = train_test_split(updated_train_set, updated_train_set["TARGET"],test_size=0.2)
```

```python
sc = StandardScaler()
scaled_train_x = sc.fit_transform(train_x)
scaled_test_x = sc.transform(test_x)
lg = LogisticRegression()
lg.fit(scaled_train_x,train_y)
train_preds = lg.predict(scaled_train_x)
print(f"Train ROC_AUC Accuracy : {roc_auc_score(train_y, train_preds)}")
test_preds = lg.predict(scaled_test_x)
print(f"Test ROC_AUC Accuracy : {roc_auc_score(test_y, test_preds)}")
```

```python
updated_test_set['TARGET'] = my_TARGET_train
sub_x = sc.transform(updated_test_set)
updated_test_set.drop("TARGET", axis='columns', inplace=True)

last_result = lg.predict_proba(sub_x)
last_result = last_result[: , 1]
my_submission = pd.DataFrame({'SK_ID_CURR' : my_SK_ID_test, 'TARGET' : last_result})
my_submission.to_csv('submission_2.csv', index = False)
```
